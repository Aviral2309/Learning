```python
import numpy as np
import pandas as pd
my_df = pd.read_csv('csv_sample.csv')
my_df
```

```python
# add column from list
gender = ['Male','Female','Male','Female','Female']
my_df["Gender"] = gender
my_df
```

```python
# add default values
my_df['Alive/Dead'] = ['jinda']*len(my_df)
my_df
```

```python
# add nan value
my_df['Show name'] = [np.nan]*len(my_df)
my_df
```

```python
# add column with .insert() - allows position
my_df.insert(1,'Adopted', [True]*len(my_df), True)
my_df
```

```python
# add column wiht .assign() - creates new df
my_df2 = my_df.assign(Relation = [False]*len(my_df))
my_df2
```
